### 1. Bernoulli Distribution
X ~ Bernoulli(p)
* It has only 1 trial
* The outcome is either 0 or 1
* $P(x) = p^{x} * q^{(1-x)}$


```python
bernoulli_data = bernoulli.rvs(size = 5000, p =0.3) #size -  number of random numbers to generate
bernoulli_data[:20]
```

```python
viz_bernoulli = sns.displot(bernoulli_data,
                             kde =False,
                             color="skyblue", alpha = 1)
viz_bernoulli.set(xlabel='Number of Success', ylabel='Frequency', title = "Histogram of Bernoulli Distribution")
```

# Q1: If the probability of finding a defective item in a certain type of good is 0.7. If X = 1 indicates probability of a defective item and X = 0 indicates a non- defective item. Find the random variate for 20 times.

```python
defectives = bernoulli.rvs(size = 20, p = 0.7)
viz_defectives = sns.displot(defectives,
                             kde = False,
                             color="orange", alpha = 1)
viz_defectives.set(xlabel='Number of Success', ylabel='Frequency', title ="Histogram of Bernoulli Distribution")
```

### Finding the mean and variance

```python
mean =defectives.mean()
variance = defectives.var()
print(f"The average number of defectives is: , {mean:.2f}")
print(f"The variance observed in the defectives is: , {variance:.2f}")
```

### 2. Binomial Distribution
X ~ Binomial(n,p), # n: number of trials and p: probability of success.
* Trails are independent
* Probability of success and failure remains the same for all trials
* $P(x) =  C_n^x * p^{x} * q^{n-x}$







```python
from scipy.stats import binom
```

```python
binomial_data = binom.rvs(n =  20, p=0.3, size = 5000)
binomial_data[:10]
```

```python
viz_binomial = sns.displot(binomial_data,
                             kde =False,
                             color="skyblue", alpha = 1)
viz_binomial.set(xlabel='Number of Success', ylabel='Frequency', title = "Histogram of Binomial Distribution")
```

## for a binomial distribution with n = 20 trials and  p =0.6, what is the probability of getting greater 15 successes?

```python
n = 20
p = 0.6
prob_greater_15 = 1 - binom.cdf(15, n,p)
print(f"The probability of getting greater than 15 successes is: {prob_greater_15:.4f}")
```

## What is the probability of getting exactly 8 successes?

```python
n = 20
p = 0.6
prob_exactly_8 = binom.pmf(8,n,p)
print(f"The probability of getting exactly 8 successes is: {prob_exactly_8:.4f}")
```

## 3. Poisson Distribution

X ~ P(λ) where λ is the rate parameter.
* $P(x) =  e^{-λ} {λ}^x /  x!$
* The number of trials "n" tends to infinity
* Probability of success "p" tends to zero
* np = 1 is finite


```python
from scipy.stats import poisson
```

```python
poisson_data = poisson.rvs(mu = 5, size = 10000 )
poisson_data
```

```python
viz_poisson = sns.displot(poisson_data,
                             kde =False,
                             color="skyblue", alpha = 1)
viz_poisson.set(xlabel='Number of Success', ylabel='Frequency', title = "Histogram of Poisson Distribution")
```

Find the probability that the user sends less than 500 messages each day.

## 5. Uniform Distribution
* Probability of each outcome is equally likely.
* X ~ U(a,b)
* $f(x) = 1/(b-a) ; a <x<b$

```python
from scipy.stats import uniform
```

```python
uniform_data = uniform.rvs(size = 5000, loc = 0, scale = 50)
uniform_data[:10]
```

```python
viz_uniform = sns.displot(uniform_data,
                          kde =False,
                          color ="skyblue",
                          alpha = 1)
viz_uniform.set(xlabel  = "Number of Success", ylabel="Frequency")
```

## Q: The waiting time (in minutes) for a bus at a certain bus stop follows a uniform distribution between 5 and 15 minutes. Let X be the random variable representing the waiting time.
* Find the probability that a randomly selected person waits less than 10 minutes for the bus.
* Calculate the expected value (mean) and variance of the waiting time.

```python
a = 5
b = 15
prob_less_than_10 = uniform.cdf(10, loc=a, scale = b-a)
print(f"1. Probability of waiting less than 10 minutes: {prob_less_than_10:.2f}")
```

```python
mean_waiting_time = uniform.mean(loc=a, scale=b-a)
variance_waiting_time = uniform.var(loc=a, scale=b-a)
print(f"2. Mean waiting time: {mean_waiting_time:.2f}")
print(f"   Variance of waiting time: {variance_waiting_time:.2f}")
```

## Q: In a normal distribution with mean =10 and standard deviation = 2. what is the probability that a randomly selected value is greater than 12?


```python
mean_value = 10
std_deviation = 2

prob_greater_12 = 1 - norm.cdf(12, loc=mean_value, scale=std_deviation)

print(f"The probability of a randomly selected value being greater than 12 is: {prob_greater_12:.4f}")
```

## 8. How does the paramters mean and standard deviation affect the normal distribution?

```python
mean_values = [5, 10, 15]
std_deviation_values = [1, 2, 3]
x_values = np.linspace(-10, 30, 1000)


plt.figure(figsize=(12, 6))

for mean, std_dev in zip(mean_values, std_deviation_values):
    pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)
    plt.plot(x_values, pdf_values, label=f'Mean={mean}, SD={std_dev}')

plt.title('Normal Distributions with Varying Mean and Standard Deviation')
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

* The highest peak refers to the mean.
* As the mean increases, the entire distribution shifts towards the right.
* As the SD increases, the width of the distribution also increases.



### 9. Naive Bayes Classifier

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
from sklearn.metrics import confusion_matrix
```

```python
#reading dataset
Data=pd.read_csv('Social_Network_Ads.csv')
print(Data.head(10))
Data.describe()
```

```python
#training and testing set size
train_size=int(0.75*Data.shape[0])
test_size=int(0.25*Data.shape[0])
print("Training set size : "+ str(train_size))
print("Testing set size : "+str(test_size))
```

```python
#Getting features from dataset
Data=Data.sample(frac=1)
X=Data.iloc[:,[2, 3]].values
y=Data.iloc[:,4].values
X=X.astype(float)
```

```python
#training set split
X_train=X[0:train_size,:]
y_train=y[0:train_size]
#testing set split
X_test=X[train_size:,:]
y_test=y[train_size:]
```

```python
#visualize the training set
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j,marker='.')
plt.title('Training set')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
cm2=confusion_matrix(y_test,y_pred)
print(cm2)
```

### 10. Birth day Problem.How many people are required so that any two people in the group have the same birthday with at least a 50-50 chance?

It is easier to calculate the inverse of the problem. That is, the probability that two people in a group do not have the same birthday. We can then invert the final result to give the desired probability, for example:

```python
# define maximum group size
n = 30
# number of days in the year
days = 365
# calculate probability for different group sizes
p = 1.0
for i in range(1, n):
  av = days - i
  p *= av / days
  print('n=%d, %d/%d, p=%.3f 1-p=%.3f' % (i+1, av, days, p*100, (1-p)*100))
```

## Alternataively we can also use numpy module to generate the distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, geom, norm, uniform, expon
```

```python
np.random.seed(42)
num_samples = 1000
```

```python
# Bernoulli distribution
p_bernoulli = 0.3
data_bernoulli = bernoulli.rvs(p_bernoulli, size=num_samples)

# Binomial distribution
n_binomial = 10
p_binomial = 0.4
data_binomial = binom.rvs(n_binomial, p_binomial, size=num_samples)

# Poisson distribution
lambda_poisson = 5
data_poisson = poisson.rvs(lambda_poisson, size=num_samples)

# Geometric distribution
p_geometric = 0.2
data_geometric = geom.rvs(p_geometric, size=num_samples)

# Normal distribution
mean_normal = 0
std_normal = 1
data_normal = norm.rvs(mean_normal, std_normal, size=num_samples)

# Uniform distribution
low_uniform = 0
high_uniform = 1
data_uniform = uniform.rvs(low_uniform, high_uniform, size=num_samples)

# Exponential distribution
lambda_exponential = 0.5
data_exponential = expon.rvs(scale=1/lambda_exponential, size=num_samples)
```

```python


# Plot histograms
plt.figure(figsize=(15, 10))

# Bernoulli distribution
plt.subplot(2, 4, 1)
plt.hist(data_bernoulli, bins=np.arange(-0.5, 2.6, 1), edgecolor='black', align='mid') #bins-  intervals into which the data is divided.each bin will represent a range of values.
plt.title('Bernoulli')

# Binomial distribution
plt.subplot(2, 4, 2)
plt.hist(data_binomial, bins=np.arange(-0.5, n_binomial + 1.5, 1), edgecolor='black', align='mid')
plt.title('Binomial')

# Poisson distribution
plt.subplot(2, 4, 3)
plt.hist(data_poisson, bins=np.arange(-0.5, 15.5, 1), edgecolor='black', align='mid')
plt.title('Poisson')

# Geometric distribution
plt.subplot(2, 4, 4)
plt.hist(data_geometric, bins=np.arange(-0.5, max(data_geometric) + 1.5, 1), edgecolor='black', align='mid')
plt.title('Geometric')

# Normal distribution
plt.subplot(2, 4, 5)
plt.hist(data_normal, bins=20, edgecolor='black')
plt.title('Normal')

# Uniform distribution
plt.subplot(2, 4, 6)
plt.hist(data_uniform, bins=20, edgecolor='black')
plt.title('Uniform')

# Exponential distribution
plt.subplot(2, 4, 7)
plt.hist(data_exponential, bins=20, edgecolor='black')
plt.title('Exponential')

plt.tight_layout()
plt.show()
```
