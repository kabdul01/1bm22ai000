## creating a post fix expression

```python
def is_operator(char):
    return char in {'+', '-', '*', '/'}

def precedence(operator):
    if operator == '+' or operator == '-':
        return 1
    elif operator == '*' or operator == '/':
        return 2
    else:
        return 0

def infix_to_postfix(infix_expression):
    stack = []
    postfix_expression = []

    for char in infix_expression:
        if char.isalnum():
            # If the character is an operand, add it to the postfix expression
            postfix_expression.append(char)
        elif char == '(':
            # If the character is an open parenthesis, push it onto the stack
            stack.append(char)
        elif char == ')':
            # If the character is a closing parenthesis, pop operators from the stack
            # and append them to the postfix expression until an open parenthesis is encountered
            while stack and stack[-1] != '(':
                postfix_expression.append(stack.pop())
            stack.pop()  # Pop the open parenthesis from the stack
        elif is_operator(char):
            # If the character is an operator, pop operators from the stack and append
            # them to the postfix expression until the stack is empty or an operator with
            # lower precedence is encountered
            while stack and precedence(stack[-1]) >= precedence(char):
                postfix_expression.append(stack.pop())
            stack.append(char)

    # Pop any remaining operators from the stack and append them to the postfix expression
    while stack:
        postfix_expression.append(stack.pop())

    return ''.join(postfix_expression)
# Example usage:
#infix_expression = input("enter the infix expression")
infix_expression = input()
postfix_expression = infix_to_postfix(infix_expression)
print("Infix Expression:", infix_expression)
print("Postfix Expression:", postfix_expression)
```

```python
# Function to check if the given character is an operator
def is_operator(char):
    return char in '+-*/^'

# Function to get the precedence of the operator
def precedence(char):
    if char == '^':
        return 3
    elif char in '*/':
        return 2
    elif char in '+-':
        return 1
    else:
        return 0

# Function to convert infix expression to postfix expression
def infix_to_postfix(expression):
    postfix = ''
    stack = []

    for char in expression:
        if char.isalnum():  # If character is an operand, add it to the postfix expression
            postfix += char
        elif char == '(':   # If character is '(', push it onto the stack
            stack.append(char)
        elif char == ')':   # If character is ')', pop operators from stack and add to postfix until '(' is encountered
            top_operator = stack.pop()
            while top_operator != '(':
                postfix += top_operator
                top_operator = stack.pop()
        elif is_operator(char):  # If character is an operator
            while stack and precedence(stack[-1]) >= precedence(char):  # Pop higher or equal precedence operators from stack
                postfix += stack.pop()
            stack.append(char)  # Push the current operator onto the stack

    # Pop any remaining operators from the stack and add to postfix
    while stack:
        postfix += stack.pop()

    return postfix

# Test the function
infix_expression = input("Enter the infix expression: ")
postfix_expression = infix_to_postfix(infix_expression)
print("Postfix expression:", postfix_expression)
```

## Circular Queue

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

class CircularQueue:
  def __init__(self):
    self.front = None
    self.rear = None

  def is_empty(self):
    return self.front == None

  def is_full(self):
    return self.rear != None

  def enqueue(self,item):
    newnode = Node(item)
    if self.is_empty():
      newnode.next = newnode
      self.front= self.rear = newnode
    else:
      newnode.next = self.front
      self.rear.next = newnode
      self.rear = newnode

  def deque(self):
    if self.is_empty():
      print("Queue is empty")
    else:
      if self.front == self.rear:
        self.front = self.rear = None
      else:
        self.front = self.front.next
        self.rear.next = self.front


  def display(self):
    if self.is_empty():
      print("Queue is empty")
    else:
      current = self.front
      print('start:',end=' ')
      while True:
        print(current.data,end=' ')
        current = current.next
        if current == self.front:
          break
      print(": end",end='')


cq  = CircularQueue()
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.display()
print('\n')
cq.deque()
cq.display()
print('\n')
cq.deque()
cq.display()

```

## Binary Search Tree, Inorder, Postorder and Preorder

```python
class TreeNode:
    def __init__(self, key):
        self.val = key
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, root, value):
        if root is None:
            return TreeNode(value)
        else:
            if value < root.val:
                root.left = self.insert(root.left, value)
            else:
                root.right = self.insert(root.right, value)
        return root

    def inorder(self,root):
      if root:
        self.inorder(root.left)
        print(root.val,end=" ")
        self.inorder(root.right)

    def preorder(self,root):
      if root:
        print(root.val,end=" ")
        self.preorder(root.left)
        self.preorder(root.right)

    def postorder(self,root):
      if root:
        self.postorder(root.left)
        self.postorder(root.right)
        print(root.val,end=" ")

    def insertkey(self,key):
      self.root = self.insert(self.root,key)

BS = BST()
# BS.insertkey(100)
# BS.insertkey(90)
# BS.insertkey(80)
# BS.insertkey(110)
# BS.insertkey(120)

BS.insertkey(90)
BS.insertkey(10)
BS.insertkey(8)
BS.insertkey(45)
BS.insertkey(56)
BS.insertkey(89)

print("In-Order : ")
BS.inorder(BS.root)
print('\n')
print("Pre-Order : ")
BS.preorder(BS.root)
print('\n')
print("Post-Order : ")
BS.postorder(BS.root)
```

## AVL Trees

```python
#AVL tree
class AVLNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        self.root = self._insert(self.root, key)

    def _insert(self, node, key):
        if not node:
            return AVLNode(key)
        elif key < node.key:
            node.left = self._insert(node.left, key)
        else:
            node.right = self._insert(node.right, key)

        # Update height of the current node
        node.height = 1 + max(self._get_height(node.left), self._get_height(node.right))

        # Check balance factor and rebalance if necessary
        balance = self._get_balance(node)

        if balance > 1:
            if key < node.left.key:
                return self._rotate_right(node)
            else:
                node.left = self._rotate_left(node.left)
                return self._rotate_right(node)
        if balance < -1:
            if key > node.right.key:
                return self._rotate_left(node)
            else:
                node.right = self._rotate_right(node.right)
                return self._rotate_left(node)

        return node

    def _rotate_right(self, z):
        y = z.left
        T3 = y.right

        y.right = z
        z.left = T3

        z.height = 1 + max(self._get_height(z.left), self._get_height(z.right))
        y.height = 1 + max(self._get_height(y.left), self._get_height(y.right))

        return y

    def _rotate_left(self, z):
        y = z.right
        T2 = y.left

        y.left = z
        z.right = T2

        z.height = 1 + max(self._get_height(z.left), self._get_height(z.right))
        y.height = 1 + max(self._get_height(y.left), self._get_height(y.right))

        return y

    def _get_height(self, node):
        if not node:
            return 0
        return node.height

    def _get_balance(self, node):
        if not node:
            return 0
        return self._get_height(node.left) - self._get_height(node.right)

    def _inorder(self, node):
        if node:
            self._inorder(node.left)
            print(node.key, end=" ")
            self._inorder(node.right)

    def inorder_traversal(self):
        self._inorder(self.root)


# Example usage:
avl_tree = AVLTree()
avl_tree.insert(10)
avl_tree.insert(20)
avl_tree.insert(30)
avl_tree.insert(40)
avl_tree.insert(50)
avl_tree.insert(25)

print("Inorder traversal of AVL tree:")
avl_tree.inorder_traversal()



#Adjust Pointers: We rearrange the pointers to perform the rotation:
#in _rotate_right
#y.right becomes z: We make y the new root of the subtree, so its right child becomes the current node z.
#z.left becomes T3: The right subtree of y (T3) becomes the left child of z



#In both functions, z is the node around which the rotation is performed.
#In the left function, y is the right child of z, while in the right function, y is the left child of z.
#T2 (for left function) and T3 (for right function) represent the subtree that might be affected during rotation.
#The rotation involves changing pointers to make y the new root of the subtree and z its left or right child accordingly.
#After the rotation, heights of the affected nodes (z and y) are updated to reflect the changes made.
#Finally, the function returns the new root of the rotated subtree (y).
```

```python
import time

class DiskDrive:

  def __init__(self, size):
    self.data = bytearray(size)

  def read_sector(self, sector):
    time.sleep(0.01)
    return self.data[sector * 512: (sector + 1) * 512]

  def write_sector(self, sector, data):
    time.sleep(0.01)
    self.data[sector * 512: (sector + 1) * 512] = data

disk = DiskDrive(1024 * 1024)

data = disk.read_sector(10)
print(data)

data_to_write = bytearray([0xAB, 0xCD, 0xEF, 0x12])
disk.write_sector(20, data_to_write)
```
